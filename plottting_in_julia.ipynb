```julia
using Plots.PlotMeasures
using Plots
pyplot()
p=range(0,step=pi/10,2*pi)
q=@. sin(p)
r=range(0,step=pi/10,4*pi)
s= @. sin(r)
a=range(0,step=pi/10,6*pi)
b=@. sin(a)
c=range(0,step=pi/10,8*pi)
d=@. sin(c)
plot1=plot(p,q,legendtitle=:plot1,title="sin(p)",subplot_index=:1)
plot2=plot(r,s,legendtitle=:plot2,title="sin(q)")
plot3=plot(a,b,legendtitle=:plot3,title="sin(a)")
plot4=plot(c,d,legendtitle=:plot4,title="sin(b)")

plot(plot1,plot2,plot3,plot4,layout=grid(2,2),
    framestyle=:semi,xlabel="time",ylabel="sin(time)",
    bottom_margin=10mm,right_margin=10mm)#gridalpha=:0.1)

#for more info see: https://docs.juliaplots.org/latest/attributes/
```
