```julia
# Model
figpath = "C:/Users/pande/Desktop/Desktop/Thesis/ThesisLyx/figures/chapter3/";
using NLsolve
using DifferentialEquations
using Plots
pyplot();# pyplot attributes on backend
using LaTeXStrings;
using Compat, Random, Distributions
# Linewidths and styles
LW1 = 2 
LW2 = 1.5
LS1 = :solid
LS2 = :dot
LS3 = :dashdot
LS4 = :dash;
# inputs (need to specify first than parameters as Qdfs is input dependent and considered as parameters)


Ifd =If_float
It = It_float
Twc = Twc_float#3.8
QdFes = 212.
Wdf = 528.
mdw = 48.51#53.9#32.42+0.038*QdFes
mda = 44.28#49.2#29.6+0.035*QdFes
function u_Ifd(t)
    # Iterative value of Twc from real data
    return Ifd[t]
end
function u_It(t)
    # Iterative value of Twc from real data
    return It[t]
end
function u_Twc(t)
    # Iterative value of Twc from real data
    return Twc[t]
end



#Specifying Parameters
pa = 1.01e5 # " Atmospheric pressure , Pa "
#
chpa = 1.012#1.15 # " Specific heat capacity air , kJ.kg -1. K -1 # " 
chpw = 4.1813#4.2 # " Specific heat capacity water , kJ.kg -1. K -1 # " 
chpCu = 0.385#0.385 # " Specific heat capacity copper , kJ.kg -1. K -1 # " 
chpFe = 0.412#0.465 # " Specific heat capacity iron , kJ.kg -1. K -1 # " 
#
mr = 9260. # " Mass of copper in rotor , kg # " 
ms = 6827. # " Mass of copper in stator , kg # " 
mFe = 71200. # " Mass of iron in stator , kg # " 
#
VhCu = 0.112e-3 # " Specific volume of copper , m3 / kg # " 
VhFe = 0.127e-3 # " Specific volume of iron , m3 / kg # " 
#
Vr = mr*VhCu # " Rotor copper volume , m3 # "  

Vs = ms*VhCu # " Stator copper volume , m3 # " 
VFe = mFe*VhFe # " Stator iron volume , m3 # " 

hAax = 55.6
hAwx = 222.

UAx = 1/(1/hAax+1/hAwx)
    

# Stanton numbers
NSta = UAx/chpa/(mda)
NStw = UAx/chpw/(mdw)
NStd = NStw - NSta

#
UAr2d = 2.96#2.835  #2.7 # " Heat transfer , rotor copper to air gap , kW / K # " 
UAs2Fe = 18#19.#20. # " Heat transfer , stator copper to stator iron , kW /K" 
UAFe2a = 12.87#13.585#14.285 #14.3 # " Heat transfer , rotor iron to hot air , kW / K # " 
#Up=44.46 # "Product of overall heat transfer and perimeter"
#
Hha_o = 0. # " Enthalpy of formation air , kJ / kg " 
HhCu_o = 0. # " Enthalpy of formation copper , kJ / kg # " 
HhFe_o = 0. # " Enthalpy of formation iron , kJ / kg # " 
#
Ta_o = 25. # " Standard state temperature air , C # " 
TCu_o = 25. # " Standard state temperature copper , C # " 
TFe_o = 25. # " Standard state temperature iron , C # " 
#
Rr = 0.127e-3 # " Ohmic resistance , rotor copper , kOhm, at 25 C # " 
Rs = 1.95e-6  # " Ohmic resistance , stator copper , kOhm, at 25 C"
#
Qdfs = 0.8*Wdf
#tuples of parameters
par = [pa,chpa,chpw,chpCu,chpFe,mr,ms,mFe,VhCu,VhFe,
    Vr,Vs,VFe,UAr2d,UAs2Fe,UAFe2a,Hha_o,HhCu_o,
    HhFe_o,Ta_o,TCu_o,TFe_o,Rr,Rs,Qdfs,hAax,hAwx,UAx,NStw,NSta,NStd];
# MODEL 1: cp, Rr, Rs are Constant

# This function returns Tac, Tad, Tah
function find_air_temperatures(temp_of_states, par;iterator=iterator)
    Tr, Ts, TFe= temp_of_states
    #print(iterator)
    function f!(F, x)
        Tac,Tad,Tah = x[1], x[2], x[3]
        F[1] = (mda)*chpa*(Tac-Tad)+UAr2d*(Tr-Tad)+Qdfs
        F[2] = (mda)*chpa*(Tad-Tah)+UAFe2a*(TFe-Tah)
        F[3] = Tac*(NStw-NSta*exp(-NStd))-NStd*Tah-NSta*(1-exp(-NStd))*u_Twc(iterator)
    end
    return nlsolve(f!,[Ta_o,Ta_o,Ta_o]).zero   
end

#This function returns ODE
function process_f(x_state, iterator)
    Trotor, Tstator, Tiron = x_state
    #print("inside ODE")
    Taircold, Tairdelta, Tairhot = find_air_temperatures([Trotor, Tstator, Tiron], par;iterator=iterator)#[Tr0, Ts0, TFe0]
    Tr=Trotor + dt*(1.1*Rr*u_Ifd(iterator)^2-UAr2d*(Trotor-Tairdelta))/(mr*chpCu)
    Ts=Tstator + dt*(3*Rs*u_It(iterator)^2-UAs2Fe*(Tstator-Tiron))/(ms*chpCu)
    TFe=Tiron + dt*(UAs2Fe*(Tstator-Tiron)-UAFe2a*(Tiron-Tairhot)+QdFes)/(mFe*chpFe)
    return Taircold, Tairdelta, Tairhot, Tr, Ts, TFe
end
```

```julia
# UKF from Dan Simon: Optimal state estimation book
using LinearAlgebra
#process noise
wk=1e-3*[9.93,2.86,3.62]
# Process noise covariance
W = 1e-6*[9.93^2 0 0 
    0 2.86^2 0
    0 0 3.62^2]
# Measurement noise
vk=1*[0.5,0.5]
# Measurement noise covariance
V = 1*[0.25 0
    0 0.25]
# Initial guess
xhatkminus1plus=[28.,28.,28.]
x = xhatkminus1plus
# Setting initial state covariance to process covariance
Pkminus1plus=W
# Simulation step
dt=60
# Simulation time range
t_sim=range(0,stop=583*60,step=dt)
# Creating variables to store data for later plottings
Tsmeas, TFemeas=zeros(0),zeros(0),zeros(0)
Trest, Tsest, TFeest=zeros(0),zeros(0),zeros(0)
Trsim, Tssim, TFesim = zeros(0),zeros(0),zeros(0)
# fucntion that calcualte xhatk
function xhatkminus(Pk,xhatkminus,iterator)
    # sigma points
    if iterator<4
        xtilde = cholesky(3*Pk).U[iterator,:]
    elseif iterator>3
        xtilde = -cholesky(3*Pk).U[iterator-3,:]
    end
    xhatk =  xhatkminus + xtilde
    
    return xhatk
end
# Loop starts k=2,3,...
for (index,value) in enumerate(t_sim)
    # Simulate real system
    Tr0, Ts0, TFe0 = x
    Tac, Tad, Tah, Tr, Ts, TFe=process_f([Tr0, Ts0, TFe0],index)
    x=[Tr,Ts,TFe]+wk*randn()
    append!(Trsim,x[1]);append!(Tssim,x[2]);append!(TFesim,x[3])
    yk=[x[2];x[3]]+vk*randn()
    append!(Tsmeas,yk[1])
    append!(TFemeas,yk[2])
    # Auxilliary matrices in UKF
    xhatminuskmul2n=zeros(3,1)
    yhatminuskmul2n=zeros(2,1)
    xhatk3by6=zeros(3,6)# creating empty for storing all sigma points
    yhatk2by6=zeros(2,6)
    Pminusk=zeros(3,3)
    Pxy=zeros(3,2)
    Py=zeros(2,2)
    # Loop for propagation
    for j=1:(2*length(xhatkminus1plus))
        Tr0, Ts0, TFe0 = xhatkminus(Pkminus1plus,xhatkminus1plus,j)
        Tac, Tad, Tah, Tr, Ts, TFe=process_f([Tr0, Ts0, TFe0],index)
        xhatk=[Tr,Ts,TFe]
        yhatk=[xhatk[2];xhatk[3]] 
        xhatk3by6[:,j] = xhatk
        yhatk2by6[:,j] = yhatk
        xhatminuskmul2n= xhatminuskmul2n+xhatk
        yhatminuskmul2n= yhatminuskmul2n+yhatk
    end
    # a priori estimate
    xhatminusk=xhatminuskmul2n/(6)
    yhatminusk=yhatminuskmul2n/(6)
    
    for k=1:6
        Pminusk=Pminusk+(xhatk3by6[:,k]-xhatminusk)*transpose(xhatk3by6[:,k]-xhatminusk)
        Py=Py+(yhatk2by6[:,k]-yhatminusk)*transpose(yhatk2by6[:,k]-yhatminusk)
        Pxy=Pxy+(xhatk3by6[:,k]-xhatminusk)*transpose(yhatk2by6[:,k]-yhatminusk)
    end
    Pminusk=Pminusk/6+W
    # Covariances and cross covariances
    Py=Py/6+V
    Pxy=Pxy/6
    # Kalman gain
    Kk=Pxy*inv(Py)
    # Innovation
    innovation=yk-yhatminusk
    xhatkminus1plus= xhatminusk+Kk*innovation
    # Storing estimate values
    xhatkminus1plus=xhatkminus1plus
    append!(Trest,xhatkminus1plus[1])
    append!(Tsest,xhatkminus1plus[2])
    append!(TFeest,xhatkminus1plus[3])
    Pkminus1plus=Pminusk-Kk*Py*transpose(Kk)
    Pkminus1plus=Hermitian(Pkminus1plus)
end
# Storing in good names for later plotting
t_sim_UKF,Tr_sim_UKF,Ts_sim_UKF,TFe_sim_UKF,
Ts_meas_UKF,TFe_meas_UKF,
Tr_est_UKF,Ts_est_UKF,TFe_est_UKF=t_sim,Trsim,Tssim,TFesim,Tsmeas,TFemeas,Trest,Tsest,TFeest
using JLD
UKF_kalman_model1=[t_sim_UKF,Tr_sim_UKF,Ts_sim_UKF,TFe_sim_UKF,
Ts_meas_UKF,TFe_meas_UKF,
Tr_est_UKF,Ts_est_UKF,TFe_est_UKF]
save("UKF_kalman_model1.jld","UKF_kalman_model1",UKF_kalman_model1)
```

```julia
# Ensemble kalman filter np=10
c=[0 1 1]
#Initialization
#states: mean=x0, var=0.1^2
x0=[28.,28.,28.]
# store initial in simulation x for plotting real system
x=x0
#process noise: mean=1, var=0.1^2
wk0=0*[0.822,1.716,0.8672]
#meas noise: mean=1, var=0.1^2
vk0=0*[1.,1.]
# random points
np=10
# initial realization/particles
x00=zeros(3,np)
wk=zeros(3,np)
vk=zeros(2,np)
#Create random of x0~(mean=x0, var=X0)
# where mean(x)=[[94.,81.,70.], sd=[0.1,0.1,0.1], var=[0.1^2,.,.]]
# This function will generate random random values for w and x

#This function returns f(xk,wk)
function process_f_ensemble(x_state,w_noise,index)
    Tr0, Ts0, TFe0 = x_state
    Tac0, Tad0, Tah0 = find_air_temperatures(x_state, par;iterator =index)#[Tr0, Ts0, TFe0]
    Tr=Tr0 + dt*(1.1*Rr*u_Ifd(index)^2-UAr2d*(Tr0-Tad0))/(mr*chpCu)
    Ts=Ts0 + dt*(3*Rs*u_It(index)^2-UAs2Fe*(Ts0-TFe0))/(ms*chpCu)
    TFe=TFe0 + dt*(UAs2Fe*(Ts0-TFe0)-UAFe2a*(TFe0-Tah0)+QdFes)/(mFe*chpFe)
    xkkminus1= [Tr,Ts,TFe]+w_noise*randn()
    return xkkminus1
end
#
#
#--Alg0rithm starts--
#
#
# Initialization k=1
# calculating x00(i), wk(i),vk(i)

wk[1,:] = rand(Normal(0, 1e-6*9.93), np) 
wk[2,:] = rand(Normal(0, 1e-6*2.86), np)
wk[3,:] = rand(Normal(0, 1e-6*3.62), np)

x00[1,:] = rand(Normal(x0[1], 1e-6*9.93), np) 
x00[2,:] = rand(Normal(x0[1], 1e-6*2.86), np)
x00[3,:] = rand(Normal(x0[1], 1e-6*3.62), np)

vk[1,:] = rand(Normal(0, 0.5), np) 
vk[2,:] = rand(Normal(0, 0.5), np)

xhat00mulnp=zeros(3,1)
X00mulnpminus1=zeros(3,3)
# calculate xhat00
for i=1:np
    xhat00mulnp=xhat00mulnp+x00[:,i]
end
xhat00=xhat00mulnp/np
#calculate X00
for i=1:np
    X00mulnpminus1=X00mulnpminus1+(x00[:,i]-xhat00)*transpose(x00[:,i]-xhat00)
end
X00=X00mulnpminus1/(np-1)
# k=2,3,.... starts
dt=60
t_sim=range(0,stop=583*60,step=dt)

Tsmeas, TFemeas=zeros(0),zeros(0)
Trest, Tsest, TFeest=zeros(0),zeros(0),zeros(0)
Trsim,Tssim,TFesim=zeros(0),zeros(0),zeros(0)
for (index,value) in enumerate(t_sim)
    # Simulating model
    #Tr0, Ts0, TFe0=states
    x = process_f_ensemble(x,1e-6*[9.93,2.86,3.62],index)
    append!(Trsim,x[1]);append!(Tssim,x[2]);append!(TFesim,x[3])
    # Measurement values
    yk=[x[2];x[3]]+[0.5;0.5]*randn()
    #yk=[Ts_meas_float[index];TFe_meas_float[index]]
    append!(Tsmeas,yk[1])
    append!(TFemeas,yk[2])
    
    # Propagation step
    #
    xkkminus1=zeros(3,np)
    for i=1:np
        xkkminus1[:,i]=process_f_ensemble(x00[:,i],wk[:,i],index)
    end
    xhatkkminus1mulnp=zeros(3,1)
    Xkkminus1mulnpminus1=zeros(3,3)
    #calulate xhatkkminus1
    for i=1:np
        xhatkkminus1mulnp=xhatkkminus1mulnp+xkkminus1[:,i]
    end
    xhatkkminus1=xhatkkminus1mulnp/np
    
    #calculate Xkkminus1: this is redundanat and can be ommited
    for i=1:np
        Xkkminus1mulnpminus1=Xkkminus1mulnpminus1+(xkkminus1[:,i]-xhatkkminus1)*transpose(xkkminus1[:,i]-xhatkkminus1)
    end
    Xkkminus1=Xkkminus1mulnpminus1/(np-1)

    # Information update
    # Measurement and innovation
    # find ykkminus1 from xkkminus1 and vk
    ykkminus1=zeros(2,np)
    for i=1:np
    # measured state is Tr
        ykkminus1[:,i]=[xkkminus1[2,i],xkkminus1[3,i]]+vk[:,i]
    end
    #calculate yhatkkminus1
    yhatkkminus1mulnp=zeros(2,1)
    for i=1:np
        yhatkkminus1mulnp=yhatkkminus1mulnp+ykkminus1[:,i]
    end
    yhatkkminus1=yhatkkminus1mulnp/np

    # Cross and innovation covariance
    #calculate Zkkminus1
    Zkkminus1mulnpminus1=zeros(3,2)
    ekkminus1mulnpminus1=zeros(2,2)
    for i=1:np
        Zkkminus1mulnpminus1= Zkkminus1mulnpminus1+(xkkminus1[:,i]-xhatkkminus1)*transpose(ykkminus1[:,i]-yhatkkminus1)
    end
    Zkkminus1=Zkkminus1mulnpminus1/(np-1)
    # calculate innovation covariance
    for i=1:np
        ekkminus1mulnpminus1= ekkminus1mulnpminus1+(ykkminus1[:,i]-yhatkkminus1)*transpose(ykkminus1[:,i]-yhatkkminus1)
    end
    ekkminus1=ekkminus1mulnpminus1/(np-1)

    # Compute Kalman gain
    Kk=Zkkminus1*inv(ekkminus1)

    # a posteriori state esitmate
    # store in initializer x00
    for i=1:np
        x00[:,i]=xkkminus1[:,i]+Kk*(yk-ykkminus1[:,i])
    end
    # store in initializer xhat00
    xhatkkmulnp=zeros(3,1)
    for i=1:np
        xhatkkmulnp=xhatkkmulnp+x00[:,i]
    end
    xhat00=xhatkkmulnp/np
    print(xhat00)
    print("\n")
    # store in X00
    Xkkmulnpminus1=zeros(3,3)
    for i=1:np
        Xkkmulnpminus1=Xkkmulnpminus1+(x00[:,i]-xhat00)*transpose(x00[:,i]-xhat00)
    end
    X00=Xkkmulnpminus1/(np-1)
    # for plotting
    
    append!(Trest,xhat00[1])
    append!(Tsest,xhat00[2])
    append!(TFeest,xhat00[3])
end
# Storing in good names for later plotting in JLD packages
t_sim_EnKF10,Tr_sim_EnKF10,Ts_sim_EnKF10,TFe_sim_EnKF10,
Ts_meas_EnKF10,TFe_meas_EnKF10,
Tr_est_EnKF10,Ts_est_EnKF10,TFe_est_EnKF10=t_sim,Trsim,Tssim,TFesim,Tsmeas,TFemeas,Trest,Tsest,TFeest
using JLD
EnKF10_kalman_model1=[t_sim_EnKF10,Tr_sim_EnKF10,Ts_sim_EnKF10,TFe_sim_EnKF10,
Ts_meas_EnKF10,TFe_meas_EnKF10,
Tr_est_EnKF10,Ts_est_EnKF10,TFe_est_EnKF10]
save("EnKF10_kalman_model1.jld","EnKF10_kalman_model1",EnKF10_kalman_model1)# Saves in the same directory
```

```julia
# np=50
# Ensemble kalman filter
c=[0 1 1]
#Initialization
#states: mean=x0, var=0.1^2
x0=[28.,28.,28.]
# store initial in simulation x for plotting real system
x=x0
#process noise: mean=1, var=0.1^2
wk0=0*[0.822,1.716,0.8672]
#meas noise: mean=1, var=0.1^2
vk0=0*[1.,1.]
# random points
np=50
# initial realization/particles
x00=zeros(3,np)
wk=zeros(3,np)
vk=zeros(2,np)
#Create random of x0~(mean=x0, var=X0)
# where mean(x)=[[94.,81.,70.], sd=[0.1,0.1,0.1], var=[0.1^2,.,.]]
# This function will generate random random values for w and x

#This function returns f(xk,wk)
function process_f_ensemble(x_state,w_noise,index)
    Tr0, Ts0, TFe0 = x_state
    Tac0, Tad0, Tah0 = find_air_temperatures(x_state, par;iterator =index)#[Tr0, Ts0, TFe0]
    Tr=Tr0 + dt*(1.1*Rr*u_Ifd(index)^2-UAr2d*(Tr0-Tad0))/(mr*chpCu)
    Ts=Ts0 + dt*(3*Rs*u_It(index)^2-UAs2Fe*(Ts0-TFe0))/(ms*chpCu)
    TFe=TFe0 + dt*(UAs2Fe*(Ts0-TFe0)-UAFe2a*(TFe0-Tah0)+QdFes)/(mFe*chpFe)
    xkkminus1= [Tr,Ts,TFe]+w_noise*randn()
    return xkkminus1
end
#
#
#--Alg0rithm starts--
#
#
# Initialization k=1
# calculating x00(i), wk(i),vk(i)
#for i=1:np
    #x00[:,i]=random_xw(x0;sd=[0.822,1.716,0.8672])
    #wk[:,i]=random_xw(wk0;sd=[0.822,1.716,0.8672])
    #vk[:,i]=random_v(vk0;sd=0.5)
#end
wk[1,:] = rand(Normal(0, 1e-3*0.822), np) 
wk[2,:] = rand(Normal(0, 1e-3*1.716), np)
wk[3,:] = rand(Normal(0, 1e-3*0.8672), np)

x00[1,:] = rand(Normal(x0[1], 1e-3*0.822), np) 
x00[2,:] = rand(Normal(x0[1], 1e-3*1.716), np)
x00[3,:] = rand(Normal(x0[1], 1e-3*0.8672), np)

vk[1,:] = rand(Normal(0, 0.5), np) 
vk[2,:] = rand(Normal(0, 0.5), np)

xhat00mulnp=zeros(3,1)
X00mulnpminus1=zeros(3,3)
# calculate xhat00
for i=1:np
    xhat00mulnp=xhat00mulnp+x00[:,i]
end
xhat00=xhat00mulnp/np
#calculate X00
for i=1:np
    X00mulnpminus1=X00mulnpminus1+(x00[:,i]-xhat00)*transpose(x00[:,i]-xhat00)
end
X00=X00mulnpminus1/(np-1)
# k=2,3,.... starts
dt=60
t_sim=range(0,stop=583*60,step=dt)

Tsmeas, TFemeas=zeros(0),zeros(0)
Trest, Tsest, TFeest=zeros(0),zeros(0),zeros(0)
Trsim,Tssim,TFesim=zeros(0),zeros(0),zeros(0)
for (index,value) in enumerate(t_sim)
    # Simulating model
    #Tr0, Ts0, TFe0=states
    x = process_f_ensemble(x,1e-3*[0.822,1.716,0.8672],index)
    append!(Trsim,x[1]);append!(Tssim,x[2]);append!(TFesim,x[3])
    # Measurement values
    yk=[x[2];x[3]]+[0.5;0.5]*randn()
    append!(Tsmeas,yk[1])
    append!(TFemeas,yk[2])
    #yk=[Ts_meas_float[index];TFe_meas_float[index]]
    
    # Propagation step
    #
    xkkminus1=zeros(3,np)
    for i=1:np
        xkkminus1[:,i]=process_f_ensemble(x00[:,i],wk[:,i],index)
    end
    xhatkkminus1mulnp=zeros(3,1)
    Xkkminus1mulnpminus1=zeros(3,3)
    #calulate xhatkkminus1
    for i=1:np
        xhatkkminus1mulnp=xhatkkminus1mulnp+xkkminus1[:,i]
    end
    xhatkkminus1=xhatkkminus1mulnp/np
    
    #calculate Xkkminus1: this is redundanat and can be ommited
    for i=1:np
        Xkkminus1mulnpminus1=Xkkminus1mulnpminus1+(xkkminus1[:,i]-xhatkkminus1)*transpose(xkkminus1[:,i]-xhatkkminus1)
    end
    Xkkminus1=Xkkminus1mulnpminus1/(np-1)

    # Information update
    # Measurement and innovation
    # find ykkminus1 from xkkminus1 and vk
    ykkminus1=zeros(2,np)
    for i=1:np
    # measured state is Tr
        ykkminus1[:,i]=[xkkminus1[2,i],xkkminus1[3,i]]+vk[:,i]
    end
    #calculate yhatkkminus1
    yhatkkminus1mulnp=zeros(2,1)
    for i=1:np
        yhatkkminus1mulnp=yhatkkminus1mulnp+ykkminus1[:,i]
    end
    yhatkkminus1=yhatkkminus1mulnp/np

    # Cross and innovation covariance
    #calculate Zkkminus1
    Zkkminus1mulnpminus1=zeros(3,2)
    ekkminus1mulnpminus1=zeros(2,2)
    for i=1:np
        Zkkminus1mulnpminus1= Zkkminus1mulnpminus1+(xkkminus1[:,i]-xhatkkminus1)*transpose(ykkminus1[:,i]-yhatkkminus1)
    end
    Zkkminus1=Zkkminus1mulnpminus1/(np-1)
    # calculate innovation covariance
    for i=1:np
        ekkminus1mulnpminus1= ekkminus1mulnpminus1+(ykkminus1[:,i]-yhatkkminus1)*transpose(ykkminus1[:,i]-yhatkkminus1)
    end
    ekkminus1=ekkminus1mulnpminus1/(np-1)

    # Compute Kalman gain
    Kk=Zkkminus1*inv(ekkminus1)

    # a posteriori state esitmate
    # store in initializer x00
    for i=1:np
        x00[:,i]=xkkminus1[:,i]+Kk*(yk-ykkminus1[:,i])
    end
    # store in initializer xhat00
    xhatkkmulnp=zeros(3,1)
    for i=1:np
        xhatkkmulnp=xhatkkmulnp+x00[:,i]
    end
    xhat00=xhatkkmulnp/np
    print(xhat00)
    print("\n")
    # store in X00
    Xkkmulnpminus1=zeros(3,3)
    for i=1:np
        Xkkmulnpminus1=Xkkmulnpminus1+(x00[:,i]-xhat00)*transpose(x00[:,i]-xhat00)
    end
    X00=Xkkmulnpminus1/(np-1)
    # for plotting
    
    append!(Trest,xhat00[1])
    append!(Tsest,xhat00[2])
    append!(TFeest,xhat00[3])
end
# Storing in good names for later plotting
t_sim_EnKF50,Tr_sim_EnKF50,Ts_sim_EnKF50,TFe_sim_EnKF50,
Ts_meas_EnKF50,TFe_meas_EnKF50,
Tr_est_EnKF50,Ts_est_EnKF50,TFe_est_EnKF50=t_sim,Trsim,Tssim,TFesim,Tsmeas,TFemeas,Trest,Tsest,TFeest
using JLD
EnKF50_kalman_model1=[t_sim_EnKF50,Tr_sim_EnKF50,Ts_sim_EnKF50,TFe_sim_EnKF50,
Ts_meas_EnKF50,TFe_meas_EnKF50,
Tr_est_EnKF50,Ts_est_EnKF50,TFe_est_EnKF50]
save("EnKF50_kalman_model1.jld","EnKF50_kalman_model1",EnKF50_kalman_model1)
```
