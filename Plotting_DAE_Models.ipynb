```julia
# Heat exchanger profile plots
# 10 min difference data of Tac, Tad, Tah
# We need Tah for heat excahnger profile
using DifferentialEquations
using BoundaryValueDiffEq
using OrdinaryDiffEq
using Plots
pyplot();
using LaTeXStrings;
#parameters for BVProblem
#p=[UP,mda,mdw,aa,ba,aw,bw,chpa,chpw]
ploting=plot(1,1,lc=:red,la=1,lw=1.5,label=L"$T_a^\mathrm{c}$")
plot!(1,1,lc=:blue,la=1,lw=1.5,label=L"$T_w^\mathrm{h}$")
# For gradient of temperatures
hot_color = range(1,length=length(Tah_ODE_co_current_model1),stop=0.2)
cold_color = range(0.2,length=length(Tah_ODE_co_current_model1),stop=1)
for (index, Tah) in enumerate(Tah_ODE_co_current_model1)
    p=[0.88,49.2,53.9,3.42,1.21e-04,3.86,4.01e-04,1.15,4.2]
    #length span of heat exchanger
    xspan=(0.0,50.0)
    function heatExchanger!(dT,T,p,time)
        #Note that heatexchanger model of bernt lie and that of
        #project description x-direction is opposite so we neglect - sign in the formula here
        #T[1]=Ta
        #T[2]=Tw
        dT[1]=p[1]/(p[2]*(p[8]))*(T[2]-T[1])
        dT[2]=p[1]/(p[3]*(p[9]))*(T[2]-T[1])
        #dT[3]=p[1]*(T[2]-T[1])
    end

    ##initialGuess=[Tah,Twh]
    initialGuess=[42.27,3.8]
    function bc!(residual,T,p,time)
        residual[1]=T[1][1]-Tah#Tah
        residual[2]=T[1][2]-3.8#Twc
        #residual[3]=T[1][3]+1125.64#Twc
    end

    bvp=BVProblem(heatExchanger!,bc!,initialGuess,xspan,p)
    sol=solve(bvp, Shooting(Vern7()))
    plot!(sol.t,[sol.u[i][1] for i in 1:length(sol.t)],la=hot_color[index],lc=:red,lw=2,label="" )#smooth=false
    ploting=plot!(sol.t,[sol.u[i][2] for i in 1:length(sol.t)],la=cold_color[index],lc=:blue,lw=2, label="", reuse=false)
end
plot!(ylabel=L"$T$ [${}^\circ$C]",xlabel=L"L_x\ [m]" );
plot!(title="Heat exchanger profile for Model 1",legend=:topright)
plot!(xlim=(0,50),reuse=false, frame=:box,size=(500,500))
display(ploting)
```
